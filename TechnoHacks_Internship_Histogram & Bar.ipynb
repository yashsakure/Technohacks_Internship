```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iris_internship.csv')
```

```python
#Create a Histogram
plt.figure(figsize=(10, 6))
plt.hist(df['SepalLengthCm'], bins=30, color='Green', edgecolor='black')  # Replace 'ColumnName' with your column
plt.title('Histogram of SepalLengthCm')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['SepalWidthCm'], bins=30, color='blue', edgecolor='black')  # Replace 'ColumnName' with your column
plt.title('Histogram of SepalWidthCm')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a Bar Chart
category_counts = df['SepalLengthCm'].value_counts()  # Replace with your categorical column

plt.figure(figsize=(10, 5))
category_counts.plot(kind='bar', color='green', edgecolor='black')
plt.title('Bar Chart of SepalLengthCm')
plt.xlabel('Categories')
plt.ylabel('Frequency')
plt.xticks(rotation=50)
plt.show()
```
